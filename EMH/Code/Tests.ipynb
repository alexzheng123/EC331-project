```python
import pandas as pd
import numpy as np
```

```python
data_path = "/Users/alexzheng/Developer/GitHub/EC331-project/EMH/Data/Bitcoin_data_2024.csv"
df = pd.read_csv(data_path, index_col=0, parse_dates=True)
df = df[['Close']]
```

```python
# Calculate returns
df['Returns'] = np.log(df['Close'] / df['Close'].shift(1))*100
df = df.dropna()

df
```

```python
# Ljung-Box Test
from statsmodels.stats.diagnostic import acorr_ljungbox

# Perform Ljung-Box test
lb_test = acorr_ljungbox(df['Returns'].dropna(), lags=[10], return_df=True)
print(lb_test)
```

```python
# Runs Test
from statsmodels.sandbox.stats.runs import runstest_1samp

# Perform Runs Test
z_stat, p_value = runstest_1samp(df['Returns'].dropna(), cutoff='mean')
print(f'Z-statistic: {z_stat}, P-value: {p_value}')
```

```python
# Variance Ratio Test
from arch.unitroot import VarianceRatio

# Perform Variance Ratio Test
vr_test = VarianceRatio(df['Returns'].dropna(), lags=10)
print(vr_test.summary())
```

```python
# Durbin-Watson Test
from statsmodels.stats.stattools import durbin_watson

# Perform Durbin-Watson Test
dw_stat = durbin_watson(df['Returns'].dropna())
print(f"Durbin-Watson Statistic: {dw_stat}")
```
