```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import talib
```

```python
df = pd.read_csv("/Users/alexzheng/Developer/GitHub/EC331-project/data/4.0/Bitcoin_data.csv",
                           index_col=0,
                           parse_dates=True)
```

```python
df.rename(columns={
    'open': 'Open',
    'high': 'High',
    'low': 'Low',
    'close': 'Close',
    'Volume BTC': 'VolumeBTC',
    'Volume USD': 'VolumeUSD'
}, inplace=True)
```

## Compute the 10 important technical indicators

```python
# Stochastics Oscillator (Fast %K and %D)
# TA-Lib STOCH returns slowk, slowd by default
df['stoch_k'], df['stoch_d'] = talib.STOCH(
    df['High'], 
    df['Low'], 
    df['Close'], 
    fastk_period=14, 
    slowk_period=3, 
    slowk_matype=0, 
    slowd_period=3, 
    slowd_matype=0
)
```

```python
# Momentum Indicator
df['momentum'] = talib.MOM(df['Close'], timeperiod=10)
```

```python
# MACD (Signal Line and MACD Oscillator)
df['macd'], df['macd_signal'], _ = talib.MACD(
    df['Close'], 
    fastperiod=12, 
    slowperiod=26, 
    signalperiod=9
)
```

```python
# Volatility Ratio Oscillator
df['atr_short'] = talib.ATR(df['High'], df['Low'], df['Close'], timeperiod=10)
df['atr_long'] = talib.ATR(df['High'], df['Low'], df['Close'], timeperiod=20)
df['vol_ratio'] = (df['atr_short'] / df['atr_long']) * 100
```

```python
# Welles Wilder’s Directional Movement Index (ADX)
df['adx'] = talib.ADX(
    df['High'], 
    df['Low'], 
    df['Close'], 
    timeperiod=14
)
```
