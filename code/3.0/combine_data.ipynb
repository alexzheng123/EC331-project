```python
import pandas as pd
```

```python
close_prices = pd.read_csv("/Users/alexzheng/Developer/GitHub/EC331-project/data/3.0/Bitcoin_data.csv",
                           index_col=0,
                           parse_dates=True)
close_prices
```

```python
sentiment_scores = pd.read_csv("/Users/alexzheng/Developer/GitHub/EC331-project/data/3.0/NEW_bitcoin_scores.csv",
                               index_col=0,
                               parse_dates=True)
sentiment_scores
```

```python
# Merge price and sentiment data
df = pd.merge(close_prices, sentiment_scores, on='date', how='outer')

# Select dates
df = df.loc['2021-02-05 10:53:00':'2021-08-21 12:59:00']

# Display first few rows
df.head()
```

```python
df.count()
```

```python
print(f"The number of data missing: \n{df.isnull().sum()}")
```

```python
# Forward-fill missing sentiment scores
df['weighted_sentiment_score'] = df['weighted_sentiment_score'].ffill()
```
