```python
import pandas as pd
```

```python
close_prices = pd.read_csv("/Users/alexzheng/Developer/GitHub/EC331-project/data/Bitcoin_data.csv",
                           index_col=0,
                           parse_dates=True)
close_prices.head()
```

```python
sentiment_scores = pd.read_csv("/Users/alexzheng/Developer/GitHub/EC331-project/data/bitcoin_scores.csv",
                               index_col=0,
                               parse_dates=True)
sentiment_scores.head()
```

```python
sentiment_scores.tail()
```

```python
df = pd.merge(close_prices, sentiment_scores, on='date', how='outer')

# Select dates between 2018-05-15 07:00:00 and 2024-01-24 22:00:00
df = df.loc['2018-05-15 07:00:00':'2024-01-24 22:00:00']

df.fillna(method='ffill', inplace=True)
```
